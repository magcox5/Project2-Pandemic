### Notebook to create dataset from United Nations Population Data (Country, Pop. by Year from 2009 to 2020) and Country_Centroids Data (Country, Lat, Lon) 

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
population_data_to_load = "Data/WPP2019_POP_F01_1_TOTAL_POPULATION_BOTH_SEXES.xlsx"

# Read Population Data from "Estimates" sheet
population_data = pd.read_excel(population_data_to_load, "ESTIMATES", encoding="ISO-8859-1")
```

```python
population_data.head(20)
```

```python
# Remove rows with "NaN" in Unnamed: 2
pop_data_1 = population_data[pd.notnull(population_data["Unnamed: 2"])]
pop_data_1
```

```python
# Select columns we want to keep
pop_data_2 = pd.DataFrame(pop_data_1, columns = ['Unnamed: 2', 'Unnamed: 5', 'Unnamed: 66', 
                                                 'Unnamed: 67', 'Unnamed: 68', 'Unnamed: 69',
                                                 'Unnamed: 70','Unnamed: 71', 'Unnamed: 72',
                                                 'Unnamed: 73', 'Unnamed: 74','Unnamed: 75', 
                                                 'Unnamed: 76','Unnamed: 77',]).copy()
```

```python
pop_data_3 = pop_data_2.rename(columns={"Unnamed: 2": "Country", "Unnamed: 5": "Type", "Unnamed: 66": "2009", 
                                       'Unnamed: 67': "2010", 'Unnamed: 68': "2011", 'Unnamed: 69': "2012",
                                       'Unnamed: 70': "2013", 'Unnamed: 71': "2014", 'Unnamed: 72': "2015",
                                       'Unnamed: 73': "2016", 'Unnamed: 74': "2017", 'Unnamed: 75': "2018",
                                       'Unnamed: 76': "2019", 'Unnamed: 77': "2020"})
pop_data_3
```

```python
# Keep only world and country rows
pop_data_4 = pop_data_3.loc[pop_data_3['Type'].isin(['Country/Area','World'])]
pop_data_4
```

```python
# Save scrubbed pop file to .csv
pop_data_4.to_csv(r'Data/pop_data_2009-2019.csv', index = False, header=True)
```
