# Pandemic file scrubbing

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

```

```python

# File to Load (Remember to Change These)
swineflu_data_to_load = "Data/Pandemic (H1N1) 2009.csv"

# Read School and Student Data File and store into Pandas Data Frames
#swineflu_data = pd.read_csv(swineflu_data_to_load, encoding=”ISO-8859–1”)
swineflu_data = pd.read_csv(swineflu_data_to_load, encoding="ISO-8859-1")
```

```python
swineflu_data
```

```python
# remove spaces in column names
swineflu_data.columns = [c.replace(' ', '_') for c in swineflu_data.columns]
# Find earliest and latest dates
print(swineflu_data.Update_Time.min())
print(swineflu_data.Update_Time.max())
swineflu_data
```

```python
# Add year column to dataframe
swineflu_data['year'] = pd.DatetimeIndex(swineflu_data['Update_Time']).year
swineflu_data
```

```python
# How many occurrences of each country in dataset?
swineflu_data['Country'].value_counts()
```

```python
# Group by country and year (I think there's only one year, but just in case...)
swineflu_1 = swineflu_data.groupby(["Country", "year"], as_index = False).agg(
    {
        'Cases':'sum',    
        'Deaths': 'sum'
    }
)

swineflu_1.tail()
```

```python
# Save scrubbed swineflu file to .csv
swineflu_1.to_csv(r'Data/swineflu_2019.csv', index = False, header=True)
```

```python
# File to Load (Remember to Change These)
covid19_data_to_load = "Data/clean_covid19_data.csv"

# Read School and Student Data File and store into Pandas Data Frames
#covid19_data = pd.read_csv(swineflu_data_to_load, encoding=”ISO-8859–1”)
covid19_data = pd.read_csv(covid19_data_to_load, encoding="ISO-8859-1")
```

```python
covid19_data
```

```python
# Add year column to dataframe
covid19_data['year'] = pd.DatetimeIndex(covid19_data['date']).year
covid19_data
```

```python
# Group by country and year (I think there's only one year, but just in case...)
covid19_data_1 = covid19_data.groupby(["country", "year"], as_index = False).agg(
    {
        'cases':'sum',    
        'deaths': 'sum'
    }
)

covid19_data_1.rename(columns={"country": "Country", "year": "Year", "cases": "Cases","deaths": "Deaths"}, errors="raise")
```

```python
covid19_data_1 = covid19_data_1.rename(columns={"country": "Country", "year": "Year", "cases": "Cases","deaths": "Deaths"}, errors="raise")
 
```

```python
covid19_data_1.columns
```

```python
# File to Load (Remember to Change These)
country_centroid_to_load = "Data/country_centroids_az8.csv"

# Read School and Student Data File and store into Pandas Data Frames
#swineflu_data = pd.read_csv(swineflu_data_to_load, encoding=”ISO-8859–1”)
country_centroid = pd.read_csv(country_centroid_to_load, encoding="ISO-8859-1")


country_centroid
```

```python
clean_country = country_centroid[["sovereignt","Longitude","Latitude"]].rename (columns={"sovereignt":"Country","Longitude":"lon","Latitude":"lat"}, errors="raise")
```

```python
clean_country.columns
```

```python
merged_covid_df = pd.merge(covid19_data_1, clean_country, how='left', on=["Country", "Country"])
```

```python
merged_covid_df
```

```python
merged_sf_df = pd.merge(swineflu_1, clean_country, how='left', on=["Country", "Country"])
```

```python
merged_sf_df.to_csv(r'Data/merged_swineflu_2019.csv', index = False, header=True)
```
