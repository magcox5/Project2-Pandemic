```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# Dependencies and Setup
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

# File to Load (Remember to Change These)
covid19_data_to_load = "Data/covid_19_data.csv"
centroid_data_to_load = "Data/country_centroids_az8.csv"
pop_data_to_load = "Data/pop_data_2009-2019.csv"

# Read School and Student Data File and store into Pandas Data Frames
#covid19_data = pd.read_csv(swineflu_data_to_load, encoding=”ISO-8859–1”)
covid19_data = pd.read_csv(covid19_data_to_load, encoding="ISO-8859-1")
centroid_data = pd.read_csv(centroid_data_to_load)
pop_data = pd.read_csv(pop_data_to_load)
```

```python
# remove spaces and '.' in column names
covid19_data.columns = [c.replace(' ', '_') for c in covid19_data.columns]
covid19_data.columns = [c.replace('.', '') for c in covid19_data.columns]


# Find earliest and latest dates
# print(covid19_data.Date.min())
# print(covid19_data.Date.max())
covid19_data
```

```python
covid19_data['Province/State'] = covid19_data['Province/State'].fillna(0)
```

```python
covid19_data_1= covid19_data.groupby(['Province/State', 'Country'], as_index = False).agg(
    {
         'Cases':"sum",    # Sum duration per group
         'Deaths': "sum",  # get the count of networks
         'Date': 'last'  # get the first date per group
    }
)
covid19_data_1
```

```python
covid19_data_1
```

```python
# Add year column to dataframe
covid19_data_1['Year'] = pd.DatetimeIndex(covid19_data_1['Date']).year
covid19_data_1
```

```python
covid_data_2 = covid19_data_1.replace(['US'], 'United States')
covid_data_2
```

```python
covid_data_3 = covid_data_2.replace(['UK'], 'United Kingdom')
covid_data_3
```

```python
covid19_data_4 = covid_data_3.groupby(['Country'], as_index = False).agg(
    {
         'Cases':"sum",    # Sum duration per group
         'Deaths': "sum",  # get the count of networks
         'Date': 'last'  # get the first date per group
    }
)
covid19_data_4
```

```python
# Merge longitude and latitude data by country
# Get name, Longitude, and Latitude from centroid_data
centroid_data_1 = centroid_data[['name', 'Longitude', 'Latitude']]
centroid_data_2 = centroid_data_1['name'].replace(['United States'], 'United States of America')

centroid_data_2
```

```python


# Merge centroid lon, lat 
covid19_2 = pd.merge(covid19_data_4 , centroid_data_1, left_on='Country', right_on='name', how = 'left')
covid19_2

```

```python
# Get subset of pop_data, Country, 2014, 2015, 2016
pop_data_1 = pop_data[['Country', '2016']]
pop_data_1
```

```python


# Merge Population Data
# Merge centroid lon, lat 
covid19_3 = pd.merge(covid19_2, pop_data_1, on='Country', how = 'left')
covid19_3

```

```python
covid19_3['population'] = covid19_3['2016'] * 1000 
covid19_3
```

```python
covid19_3.columns
```

```python
covid19_4 = covid19_3.rename(columns={"year": "Year", "Longitude": "Lon", "Latitude": "Lat" })
covid19_4
```

```python
covid19_5 = covid19_4.drop(columns=['2016'])
```

```python
covid19_5['Pandemic'] = "covid19"
covid19_5
```
