```python
covid19_data_2['Province/State'] = covid19_data_2['Province/State'].fillna(0)
```

```python
covid19_data_2.sort_values(by=['Date'])
```

```python
covid19_data_3 = covid19_data_2.groupby(['Province/State', 'Country'], as_index = False).agg(
    {
         'Cases':"sum",    # Sum duration per group
         'Deaths': "sum",  # get the count of networks
         'Date': 'last'  # get the first date per group
    }
)
covid19_data_3
```

```python
# Add year column to dataframe
covid19_data_3['Year'] = pd.DatetimeIndex(covid19_data_3['Date']).year
covid19_data_3
```

```python
covid19_data_4 = covid19_data_3.groupby(['Country'], as_index = False).agg(
    {
         'Cases':"sum",    # Sum duration per group
         'Deaths': "sum",  # get the count of networks
         'Date': 'last'  # get the first date per group
    }
)
covid19_data_4
```

```python
# Merge longitude and latitude data by country
# Get name, Longitude, and Latitude from centroid_data
centroid_data_1 = centroid_data[['name', 'Longitude', 'Latitude']]

centroid_data_1
```

```python


# Merge centroid lon, lat 
covid19_data_5 = pd.merge(covid19_data_4 , centroid_data_1, left_on='Country', right_on='name', how = 'left')
covid19_data_5

```

```python
# Get subset of pop_data, Country, 2014, 2015, 2016
pop_data_1 = pop_data[['Country', '2016']]
pop_data_1  
```

```python


# Merge Population Data
# Merge centroid lon, lat 
covid19_3 = pd.merge(covid19_data_5, pop_data_1, on='Country', how = 'left')
covid19_3

covid19_3 = covid19_3
covid19_3
```

```python
covid19_3['2016'] = covid19_3['2016'].fillna(1)
```

```python
covid19_3['population'] = covid19_3['2016'] * 1000 
covid19_3
```
