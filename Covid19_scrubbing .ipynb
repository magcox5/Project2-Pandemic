# Covid file scrubbing

```python
# Dependencies and Setup
import pandas as pd
import numpy as np


```

```python
# File to Load (Remember to Change These)
covid19_data_to_load = "Data/covid19_data.csv"

# Read School and Student Data File and store into Pandas Data Frames
#covid19_data = pd.read_csv(swineflu_data_to_load, encoding=”ISO-8859–1”)
covid19_data = pd.read_csv(covid19_data_to_load, encoding="ISO-8859-1")
```

```python
covid19_data.tail()
```

```python
# Add year column to dataframe
covid19_data['Year'] = pd.DatetimeIndex(covid19_data['Date']).year
covid19_data
```

```python
# Group by country and year (I think there's only one year, but just in case...)
covid19_data_1 = covid19_data.groupby(["Pandemic","Country", "Year"], as_index = False).agg(
    {
        'Cases':'sum',    
        'Deaths': 'sum'
    }
)

covid19_data_1 
```

```python
# covid19_data_1 .insert(0,"Pandemic", "covid19") 
```

```python
covid19_data_1 = covid19_data_1 
 
```

```python
covid19_data_1.columns
```

```python
# File to Load (Remember to Change These)
country_centroid_to_load = "Data/unused/clean_country_centroids_az8.csv"

# Read School and Student Data File and store into Pandas Data Frames
#swineflu_data = pd.read_csv(swineflu_data_to_load, encoding=”ISO-8859–1”)
country_centroid = pd.read_csv(country_centroid_to_load, encoding="ISO-8859-1")


country_centroid
```

```python
clean_country = country_centroid[["name","Longitude","Latitude"]].rename (columns={"name":"Country","Longitude":"lon","Latitude":"lat"}, errors="raise")
```

```python
clean_country.columns
```

```python
merged_covid_df = pd.merge(covid19_data_1, clean_country, how='left', on=["Country", "Country"])
```

```python
merged_covid_df
```

```python

null_columns=merged_covid_df.columns[merged_covid_df.isnull().any()]
merged_covid_df[null_columns].isnull().sum()
```

```python
print(merged_covid_df[merged_covid_df["Country"].isnull()][null_columns])
```

```python
print(merged_covid_df[merged_covid_df.isnull().any(axis=1)][null_columns].head())
```

```python
merged_covid_df.to_csv(r'Data/merged_covid_2019.csv', index = False, header=True)
```
